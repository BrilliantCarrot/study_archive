```python
time = input("현재시간: ")

# 문자열의 뒤에서 2번째부터 끝까지(-2:) 가져오면 "분"만 추출됩니다.
if time[-2:] == "00":
    print("정각 입니다.")
else:
    print("정각이 아닙니다.")
```

```python
score = int(input("학점을 입력하세요: "))
print("score: ", score)
if score <= 20:
    print("grade is E")
elif score >= 21 and score <= 40:
    print("grade is D")
elif score >= 41 and score <= 60:
    print("grade is C")
elif score >= 61 and score <= 80:
    print("grade is B")
else:
    print("grade is A")
```

```python
oddList = []
evenList = []
for i in range(1, 31):
    if i % 2 == 0:
        evenList.append(i)
    else:
        oddList.append(i)
print("홀수 리스트: ", oddList)
print("짝수 리스트: ", evenList)
```

```python
class Phone():
    def __init__(self, manufacturer = "삼성", releaseYear=2025, color = "화이트실버"):
        print("휴대폰 생성")
        self.manufacturer = manufacturer
        self.releaseYear = releaseYear
        self.color = color
        print(f"제조사: {self.manufacturer}, 출고년도: {self.releaseYear}, 색상: {self.color}")
    
my_phone = Phone()
```

```python
class Phone:
    def __init__(self, brand="미정", year=2024, color="블랙"):
        self.brand = brand
        self.year = year
        self.color = color
        print(f"제조사: {self.brand}, 출고년도: {self.year}, 색상: {self.color}")

# 1. 모든 인자를 다 넣는 경우
p1 = Phone("삼성", 2023, "실버")

# 2. 일부만 넣는 경우 (나머지는 기본값 적용)
p2 = Phone("애플") 

# 3. 아무것도 안 넣는 경우
p3 = Phone()
```
