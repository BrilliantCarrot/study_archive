```python
class Bag():
    def __init__(self, initial_items=None, color="black", bag_type="backpack"):
        if initial_items is None:
            self.items = []
        else:
            self.items = list(initial_items)
        self.color = color
        self.bag_type = bag_type
        self.is_open = False # 기본적으로 닫혀있는 상태

    def open(self):
        self.is_open = True
        print("가방을 열었습니다.")

    def close(self):
        self.is_open = False
        print("가방을 닫았습니다.")

    def add(self, x):
        if self.is_open != True:
            print("가방이 닫혀 있어서 넣을 수 없습니다.")
            return

        self.items.append(x)
        print(f"{x}을/를 넣었습니다.")

    def remove(self, x):
        if self.is_open != True:
            print("가방이 닫혀 있어서 뺄 수 없습니다.")
            return

        if x in self.items:
            self.items.remove(x)
            print(f"{x}를 빼냈습니다.")
        else:
            print(f"{x}이(은/는) 없습니다.")

    def status(self):
        item_str = ", ".join(self.items)
        print(f"가방의 색은 {self.color}이며, 가방의 종류는 {self.bag_type}이고, 가방 안에는 {item_str}가 있습니다.")

bag = Bag(["필통", "미니우산", "태블릿"],"검정색","백팩")
bag.add("블루투스 이어폰")
bag.open()
bag.add("블루투스 이어폰")
bag.remove("볼펜")
bag.close()
bag.remove("연필")
bag.status()
```
