```python
class Person:
    def __init__(self, name, age, address,wallet):
        self.hello = "안녕하세요"
        self.name = name
        self.age = age
        self.address = address
        self.__wallet = wallet # 변수 앞에 __를 붙여서 비공개(private) 속성으로 만듦
        # __이름__의 경우엔 공개(public)
    def greeting(self):
        print("{0} 저는 {1}입니다.".format(self.hello, self.name))
    def pay(self, amount):
        self.__wallet -= amount
        print("이제 {0}원 남았네요".format(self.__wallet))

maria = Person("마리아",20,"서울시 서초구 반포동")
maria.greeting()
print('이름:', maria.name)
print('나이:', maria.age)
print('주소:', maria.address)
```

```python
class Calc:
    @staticmethod
    def add(a, b):
        print(a + b)
 
    @staticmethod
    def mul(a, b):
        print(a * b)
# 정적 메서드는 self를 받지 않으므로 인스턴스 속성에 접근할 수 없다.
# 그래서 보통 정적 메서드는 인스턴스 속성, 인스턴스 메서드가 필요 없을 때 사용한다.
# 정적 메서드는 메서드의 실행이 외부 상태에 영향을 끼치지 않는 순수 함수를 만들 때 사용.
Calc.add(10, 20)    # 클래스에서 바로 메서드 호출
Calc.mul(10, 20)    # 클래스에서 바로 메서드 호출
```

```python
class Person:
    count = 0    # 클래스 속성
    def __init__(self):
        Person.count += 1    # 인스턴스가 만들어질 때
                             # 클래스 속성 count에 1을 더함
    @classmethod
    def print_count(cls):
        print('{0}명 생성되었습니다.'.format(cls.count))    # cls로 클래스 속성에 접근
 
james = Person()
maria = Person()
Person.print_count()    # 2명 생성되었습니다.
```

super(): 자식 클래스에서 부모 클래스의 속성이나 메서드에 접근할 때 사용   
특히 자식 클래스에서 __init__을 새로 만들 때, 부모가 이미 만들어둔 변수들을 그대로 가져오기위해 필수적으로 사용

```python
class Aircon():
    def __init__(self,name):
        self.name = name
        self.temperature = 24
        self.category = "스탠드형"

    def cool(self):
        print("냉방 모드를 실행합니다.")

    def sleep(self):
        print("취침 모드를 실행합니다.")

    def set_temperature(self, temperature = None):
        if temperature == None: # 만약 인자가 전달되자 않았다면 입력을 받음
            try:
                temp_input = int(input("희망 온도를 입력하세요 (18~30):"))
                if temp_input >= 18 and temp_input <= 30:
                    self.temperature = temp_input
                    print(f"{self.name}: 희망 온도가 {self.temperature}로 설정되었습니다.")
                else:
                    print("온도를 18도에서 30도 사이로 입력하세요")
            except ValueError:
                print("오류: 숫자만 입력 가능합니다.")
        else: # 2. 만약 인자가 전달되었다면 해당 인자로 온도 설정
            self.temperature = int(temperature)

    def info(self):
        print(f"이름은 {self.name}이고, 종류는 {self.category}입니다.", end=' ')

class newAircon(Aircon):
    def __init__(self, name, brand):
        super().__init__(name)
        self.brand = brand
    def warm(self):
        print("난방 모드를 실행합니다.")

    def strong(self):
        print("강풍 모드를 실행합니다.")

    def info(self):
        super().info()
        # super()를 안쓰면 밑에 코드로 실행
        # print(f"이름은 {self.name}이고, 종류는 {self.category}이고, 제조사는 {self.brand}입니다.") 
        print(f"제조사는 {self.brand}입니다.")


yourAircon = Aircon("벽걸이형 에어컨")
# yourAircon.strong() # 자식 클래스로 생성한게 아니므로 자식 클래스 함수 사용 불가능
yourAircon.set_temperature(25)
yourAircon.category = "벽걸이형"
yourAircon.info()
print("\n")
myAircon = newAircon("무풍 에어컨", "삼성")
myAircon.warm()
myAircon.set_temperature()
myAircon.strong()
myAircon.info()
```
