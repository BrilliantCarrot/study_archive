```python
path = "./pizza_file1.txt"
encoding = "utf-8"

with open(path, "r", encoding=encoding) as file:
    lines = file.readlines()

for line in lines:
    print(line, end='')
```

```python
# 앞서 생성한 'pizza_file1.txt' 파일을 읽어서 피자 종류를 리스트로 반환 및 저장하고 리스트를 출력하시오.
path = "./pizza_file1.txt"
encoding = "utf-8"

pizzaList = []
with open(path, "r", encoding=encoding) as file:
    lines = file.readlines()

    for line in lines:
        line = line.strip()
        spaceIdx = line.find(" ")
        pizzaName = line[:spaceIdx]
        pizzaList.append(pizzaName)

print(pizzaList)
```
