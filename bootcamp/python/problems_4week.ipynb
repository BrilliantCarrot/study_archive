```python
# 예외처리를 이용해 다음 조건을 만족하는 프로그램을 작성하시오.
# 1.사용자로부터 숫자를 입력받아 제곱값을 출력
# 2.숫자가 아닌 값을 입력하면 “올바른 숫자를 입력하세요!”라는 메시지를 출력하고,
# 올바른 숫자가 입력될 때 까지 반복

while True:
    try:
        num = float(input("숫자를 입력하세요: "))
        print("제곱값:", num ** 2)
        break
    except ValueError:
        print("올바른 숫자를 입력하세요!")
```

```python
# map() 함수와 람다 함수를 사용해 
# 리스트 [10, 20, 30, 40, 50]의 각 요소를 제곱한 결과를 출력하는 프로그램을 작성하시오.

numbers = [10,20,30,40,50]

squaredNumbers = map(lambda x : x**2, numbers)
print(list(squaredNumbers))
```

```python
# 주어진 문자열에서python으로 시작하는 모든 문장을 찾아 출력하는 프로그램을 작성하시오.
# data = """python one
# life is too short
# python two
# you need python
# python three"""

data = """python one
life is too short
python two
you need python
python three"""

import re

p = re.compile("^python.*", re.MULTILINE)

result = p.findall(data)

for line in result:
    print(line)
```

```python
# 사용자로부터 이메일 주소를 입력받고, 올바른 이메일 형식인지 검증하는 프로그램을 작성하시오.

import re

p = re.compile(r'^[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}$')

while True:
    email = input("이메일을 입력하세요: ")
    if p.match(email):
        print("올바른 이메일 형식입니다.")
        break
    else:
        print("이메일을 다시 입력하세요.")
```

```python
# 다음 조건을 만족하는 이터레이터 클래스를 구현하고자 한다. 빈 칸 < >에 알맞은 코드를 작성하시오.
# 1. 리스트를 인자로 받아 요소를 하나씩 반환한다.
# 2. StopIteration 예외를 적절히 처리한다.

class ListIterator:
    def __init__(self, data):
        if not isinstance(data,list):
            raise TypeError("리스트를 입력해야 합니다.")
        self.data = data
        self.index = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        if self.index < len(self.data):
            value = self.data[self.index]
            self.index += 1
            return value
        else:
            raise StopIteration
        
my_list = [1,2,3,4,5]
iterator = ListIterator(my_list)

for item in iterator:
    print(item)
```

```python
# 제너레이터 함수를 만들어 1부터 n까지 숫자의 제곱을 생성하는 프로그램을 작성하시오.

def square_generator(n):
    for i in range(1, n+1):
        yield i ** 2
for num in square_generator(5):
    print(num)
```
