```python
oroot = Tk()



oroot.geometry("400x400")

radio_value = IntVar() # 정수형 변수 생성
radio_value.set(-1) # 정수값 설정, -1은 아무것도 설정 안됨으로 설정

lunch = {0:"A런치", 1:"B런치", 2:"C런치"}

def buy():
    rValue = radio_value.get()
    print(f"{lunch[rValue]}를 구입하였습니다")


# variable: 클릭된 버튼의 정보를 저장할 변수명 설정
orbtn1 = Radiobutton(oroot, text=lunch[0], variable=radio_value, value=0)
orbtn2 = Radiobutton(oroot, text=lunch[1], variable=radio_value, value=1)
orbtn3 = Radiobutton(oroot, text=lunch[2], variable=radio_value, value=2)
orbtn1.pack()
orbtn2.pack()
orbtn3.pack()



    
obtn = Button(oroot, text="주문", command = buy)
obtn.pack()

oroot.mainloop()
```

```python
from tkinter import *

window = Tk()
window.title("조각 피자 주문 프로그램")
window.geometry("400x350")

Label(window, text="피자").pack()

cheese_var = IntVar()
combo_var = IntVar()
bulgogi_var = IntVar()

Checkbutton(window, text="치즈 피자 (3200원)", variable=cheese_var).pack(anchor="w")
Checkbutton(window, text="콤비네이션 피자 (3500원)", variable=combo_var).pack(anchor="w")
Checkbutton(window, text="불고기 피자 (3600원)", variable=bulgogi_var).pack(anchor="w")

def order():
    total = 0
    result_text = "주문내역:\n"

    if cheese_var.get() == 1:
        result_text += "- 치즈 피자 (3200원)\n"
        total += 3200

    if combo_var.get() == 1:
        result_text += "- 콤비네이션 피자 (3500원)\n"
        total += 3500

    if bulgogi_var.get() == 1:
        result_text += "- 불고기 피자 (3600원)\n"
        total += 3600

    result_text += f"\n총 가격: {total}원"

    result_label.config(text=result_text)

Button(window, text="주문", command=order).pack()
result_label = Label(window, text="")
result_label.pack()

window.mainloop()
```
