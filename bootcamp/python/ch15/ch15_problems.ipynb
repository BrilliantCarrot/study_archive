```python
# 다음 리스트를 이터레이터로 변환하고 동작을 확인하시오
food = ["김밥", "만두", "양념치킨", "족발", "피자"]
food_iter = iter(food)

print(next(food_iter))
print(next(food_iter))
print(next(food_iter))

print("-----남은 음식-----")
for item in food_iter:
    print(item)

class FoodIter:
    def __init__(self, food_list):
        self.food_list = food_list
        self.index = 0

    def __iter__(self):
        return self
    
    def __next__(self):
        # 인덱스가 리스트 길이를 넘어가면 반복 종료
        if self.index >= len(self.food_list):
            raise StopIteration
        # 현재 인덱스의 값을 가져오고 인덱스 1 증가
        result = self.food_list[self.index]
        self.index += 1
        return result
    
print("===== 클래스 동작 확인 =====")
food = ["김밥", "만두", "양념치킨", "족발", "피자"]
my_food_iter = FoodIter(food)

print(f"첫 번째: {next(my_food_iter)}")
print(f"두 번째: {next(my_food_iter)}")

print("--- 나머지 출력 ---")
for item in my_food_iter:
    print(item)
```

```python
import os
# 파일을 다음과 같이 생성 후, 파일에서 한 줄씩 읽는 제너레이터를 생성하고 사용하시오
# 제너레이터 생성 방법
# yield
# 컴프리헨션
# 클래스

data = ["김밥", "만두", "양념치킨", "족발", "피자"]
file_path = "/home/lyj/study_archive/bootcamp/food.txt"

# 1. 폴더가 없다면 생성 (실습 환경용)
os.makedirs(os.path.dirname(file_path), exist_ok=True)

# 2. 파일 쓰기 함수
def write_file(file_path, data_list):
    with open(file_path, 'w', encoding='utf-8') as file:
        for item in data_list:
            file.write(item + '\n')  # 한 줄씩 저장

# # 3. 한 줄씩 읽는 제너레이터 함수
# def read_file_generator(path):
#     with open(path, 'r', encoding='utf-8') as file:
#         for line in file:
#             yield line.strip()  # 줄바꿈 문자 제거 후 하나씩 반환

def read_file_generator(path):
    with open(path, 'r', encoding='utf-8') as file:
        lines = file.readlines()
        for line in lines:
            yield line.strip()

# 실행
if __name__ == "__main__":
    # 파일 생성
    write_file(file_path, data)
    
    # 제너레이터 생성 및 사용
    food_gen = read_file_generator(file_path)
    
    print("--- 제너레이터로 한 줄씩 읽기 ---")
    for food in food_gen:
        print(food)

type(food_gen)
```

```python
# 7. 주어진 리스트에서 next() 함수를 사용하여 각 요소를 하나씩 출력하세요. StopIteration 예외를 처리하여 출력이 끝날 때까지 반복되도록 하세요.
fruits = ["apple", "banana", "cherry"]

fruit_iter = iter(fruits)
while True:
    try:
        fruit = next(fruit_iter)
        print(fruit)
    except StopIteration:
        print("리스트 내 모든 요소를 출력하였습니다.")
        break
```

```python
# 8. 0부터 9까지의 숫자를 이터레이터로 순회하며, 각 숫자의 제곱을 출력하는 프로그램을 작성하세요.

class SquareIterator:
    def __init__(self):
        self.num = 0
        self.end = 9
    def __iter__(self):
        return self
    def __next__(self):
        if self.num > self.end:
            raise StopIteration
        result = self.num ** 2
        self.num = self.num + 1
        return result
    
square = SquareIterator()

for num in square:
    print(num)
```

```python
# 9. 0부터 10까지의 숫자 중 짝수만 출력하는 프로그램을 작성하세요. 이때, 이터레이터를 사용해야 합니다.
class EvenIterator:
    def __init__(self):
        self.num = 0
        self.end = 10
    def __iter__(self):
        return self
    def __next__(self):
        while self.num <= self.end:
            if self.num % 2 == 0:
                result = self.num
                self.num += 1
                return result
            
            self.num += 1
        raise StopIteration
    
even = EvenIterator()

for num in even:
    print(num)
```

```python
class MyRange:
    def __init__(self, start, stop, step):
        self.num = start
        self.stop = stop
        self.step = step
    def __iter__(self):
        return self
    def __next__(self):
        if self.num >= self.stop:
            raise StopIteration
        result = self.num
        self.num = self.num + self.step
        return result
        
new_range = MyRange(1,10,1)
for num in new_range:
    print(num)
```
