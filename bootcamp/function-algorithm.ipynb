```python
# 가변 위치 인자(*args): 개수가 정해지지 않은 인자
def total(*numbers):
  print(numbers)

total(1)
total(1,2)
total(1,2,3)

# 가변 키워드 인자(**kwargs): 이름 붙은 인수를 여러개 받음
def profile(**info):
  print(info)

profile(name = "홍길동", age = 553, address = "서울시")

def example(a, b = 10, *args, **kwargs):
  print("a: ", a)
  print("b: ", b)
  print("args: ", args)
  print("kwargs: ", kwargs)

example(1,2,3,4,x=100, y=200)
```

```python
def fk(cb):
  total = 0
  for sb in range(0,3,1):
    total = total + cb[sb]
  cb[2] = total
  print("함수 안 cb는: ",cb)
  return cb

ca = [10,20,30]
print("함수 전 ca는: ",ca)
cd = fk(ca)
print("함수 후 ca는: ",ca)
print("ca 리스트가 저장된 cd는: ",cd)
```

```python
# 선택 정렬 알고리즘
# 이중 for문을 통한 작성
ca = [21, 10, 11, 15, 13]

# 1. 외부 반복문: 리스트의 각 자리(i)를 순차적으로 채웁니다.
for i in range(len(ca)):
    # 현재 자리(i)의 값을 일단 최솟값으로 가정합니다.
    mina = ca[i]
    minix = i

    # 2. 내부 반복문: i번째 뒷부분(sb)에서 진짜 최솟값을 찾습니다.
    # range(i + 1, 5)는 i 다음 칸부터 끝까지 검사한다는 뜻입니다.
    for sb in range(i + 1, 5):
        if mina > ca[sb]:
            mina = ca[sb]
            minix = sb

    # 3. 데이터 교환(Swap): 찾은 최솟값(ca[minix])과 현재 자리(ca[i])를 바꿉니다.
    temp = ca[i]
    ca[i] = ca[minix]
    ca[minix] = temp

print(ca)
```

```python
# 21페이지
# 앞서 작성한 선택 정렬 부분을 fselsort 함수로 정의하고 호출하여 데이터 정렬

myList = [5, 26, 19, 41, 88]

def fselsort(list):

    for i in range(len(list)):
        
        minidx = i
        min = list[minidx]
        
        for j in range(i+1,len(list)):
            if list[j] <= min:
                min = list[j]
                minidx = j


        list[i], list[minidx] = list[minidx], list[i]


fselsort(myList)
print(myList)
```

```python
# 버블 정렬 알고리즘

numList = [5, 26, 19, 41, 88]
n = len(numList) # 리스트의 길이 (5)

# 외부 반복문: 정렬을 전체적으로 몇 번 반복할지 결정
for i in range(n):
    # 내부 반복문: 인접한 두 요소를 비교 (이미 정렬된 끝부분은 제외)
    for j in range(0, n - i - 1):
        # 앞의 숫자가 뒤의 숫자보다 크면 서로 바꿈 (오름차순)
        if numList[j] > numList[j+1]:
            # 데이터 교환(Swap)
            temp = numList[j]
            numList[j] = numList[j+1]
            numList[j+1] = temp

print(numList) # 출력: [5, 19, 26, 41, 88]
```
