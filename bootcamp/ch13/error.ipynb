```python
file_name = "myFile.txt"

try:
    f = open(file_name, 'r')
    s = f.readline()
    i = int(s.strip())
    f.close()
except OSError as err:
    print("파일이 없거나 열 수 없어 새로 생성합니다:", err)
    with open(file_name, "w") as f:
        f.write('파일 생성')
    
except ValueError:
    print("Could not convert data to an integer")
    with open(file_name, 'a') as f:
        f.write('100\n')
except Exception as err:
    print(f"Unexpected {err}, {type(err)}")
finally:
    print("프로그램 종료")
```

```python
numList = [123, 456, 789]
try:
    joinedNumList = "-".join(numList)
except TypeError:
    # 각 요소를 하나씩 문자열로 바꾼 새 리스트를 만듭니다 (리스트 컴프리헨션)
    # numList = [str(n) for n in numList]
    # map 함수, 여러 개의 데이터를 받아서, 그 안에 요소 하나하나에 똑같은 함수를 한꺼번에 적용
    # numList = list(map(str, numList))
    joinedNumList = "-".join(numList)
    strList = []
    for n in numList:
        strList.append(str(n))
    joinedNumList = "-".join(strList)
print(joinedNumList)
```

```python
# 람다 함수

adder = lambda x,y = 5 : x+y
print(adder(3))
print(adder(3,10))
print((lambda x,y : x+y)(3,10))
print((lambda x,y=5 : x+y)(3))
```

```python
numbers = [10, 20, 30]

try:
    index = int(input("인덱스를 입력하세요: "))
    print(f"해당 위치의 원소: {numbers[index]}")
except IndexError:
    print("잘못된 인덱스입니다.")
except ValueError:
    print("숫자를 입력하지 않으셨습니다.")
```
