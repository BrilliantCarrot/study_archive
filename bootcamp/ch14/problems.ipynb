```python
import re
pattern = r'(ab)+'
text = "ababab"
match = re.match(pattern, text)
print(match.group())

# 그룹화 (ab): a와 b가 순서대로 결합된 하나의 덩어리를 하나의 단위로 취급합니다.
# 반복 +: 이 (ab)라는 덩어리가 1번 이상 반복되는 것을 찾습니다.
# 동작: "ababab"는 (ab)가 총 3번 연속해서 나오므로, 전체 문자열이 패턴에 완벽히 일치하게 됩니다.
```

```python
# 문자열에서 이메일 주소를 찾아 반환

import re

text = "이메일 목록: test@example.com, hello@world.net, user123@domain.org"

pattern = r'[a-zA-Z0-9]+@[a-zA-Z0-9]+\.[a-zA-Z]{2,}'

emails = re.findall(pattern, text)

print(emails)

# [a-zA-Z0-9._%+-]+ : 사용자 아이디 (영문, 숫자, 특수문자 일부 허용)
# [a-zA-Z0-9.-]+     : 도메인 이름
# \.[a-zA-Z]{2,}    : . 뒤의 최상위 도메인 (com, net 등 2글자 이상)
# !!! 마지막 뒤에 +를 붙여야 여러 숫자 가져옴

# 결과: ['test@example.com', 'hello@world.net', 'user123@domain.org']
```

```python
text = "연락처: 010-1234-5678, 02-987-6543, 031-456-7890"

pattern = r'[0-9]+-[0-9]+-[0-9]+'

phoneNumber = re.findall(pattern, text)

print(phoneNumber)
```

```python
# 8. 주어진 문자열에서 "Python"이 포함된 문장만 리스트로 반환하세요.
# [^.]* (Python 앞뒤에 오는 내용들)
# [ ]: 대괄호는 "이 안의 문자 중 하나"를 의미합니다.
# ^: 대괄호 안에서 가장 앞에 쓰이면 "~를 제외한" (NOT) 이라는 뜻입니다.
# .: 여기서는 마침표 그 자체입니다.
# *: 0번 이상 반복.
# 즉, [^.]*는 "마침표가 아닌 문자들을 계속 가져오라"는 뜻입니다. (공백, 영문, 숫자 등 마침표만 아니면 다 통과!)


text = "I love Python. Java is also popular. Python is great for AI."
pattern = r'[^.]*Python[^.]*\.'
pythonList = re.findall(pattern, text)
print(pythonList)
```

```python
# 9.문자열에서 숫자만 추출하여 리스트로 반환하세요.

text = "상품 코드: A123, B456, C789, 가격: 12000원"
pattern = r'[0-9]+'
numbers = re.findall(pattern, text)
print(numbers)
```

```python
# 10.주어진 문자열에서 연속된 대문자(예: NASA, AI)만 찾아 리스트로 반환하세요.

text = "NASA is working on AI projects with IBM and Google."
pattern = r'[A-Z]{2,}'
capitalLatters = re.findall(pattern, text)
print(capitalLatters)
```
