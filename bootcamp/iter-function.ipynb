```python
# 덧셈 함수
def fadd(a, b):
    return a + b
# 뺄셈 함수
def fsub(a, b):
    return a - b
# 곱셈 함수
def fmul(a, b):
    return a * b
# 나눗셈 함수
def fdiv(a, b):
    if b == 0:
        return "0으로 나눌 수 없습니다!"
    return a / b
# 매개변수 설정
num1, num2 = 100, 3
# 결과 출력
print(f"덧셈 결과: {num1} + {num2} = {fadd(num1, num2)}")
print(f"뺄셈 결과: {num1} - {num2} = {fsub(num1, num2)}")
print(f"곱셈 결과: {num1} * {num2} = {fmul(num1, num2)}")
print(f"나눗셈 결과: {num1} / {num2} = {fdiv(num1, num2):.2f}")
```

```python
# 파이썬 8차시 과제 문제 6번
# 리스트에서 첫 번째 짝수와 마지막 홀수를 찾아 서로 스왑하는 프로그램을 작성하시오
a = [3,6,7,4,9,10,13]
firstEven = 0
firstEvenFlag = False
lastOdd = 0
for i in range(len(a)):
  num = a[i]
  if a[i] % 2 ==0 and firstEvenFlag == False:
    firstEven = num
    firstEvenFlag = True
  if a[i] % 2 != 0:
    lastOdd = num
print(f"첫 번째 짝수: {firstEven}")
print(f"마지막 홀수: {lastOdd}")
```

```python
# 파이썬 8차시 과제 문제 8번
# 주어진 리스트 [29, 10, 14, 37, 13]을 선택 정렬로 한 단계 수행한 결과는 무엇인가요?
data = [29, 10, 14, 37, 13]

# 0번째 자리에 들어갈 최소값 찾기
min_index = 0

for i in range(1, len(data)):
    if data[i] < data[min_index]:
        min_index = i

# 첫 번째 원소와 최소값 교환
data[0], data[min_index] = data[min_index], data[0]

print(data)
```

```python
# 파이썬 3주차 과제 문제 1번
# myList = [1,2,99,50,20,30]
# def list_max(num_list):
#   max = num_list[0]
#   for num in num_list:
#     if num >= max:
#       max = num
#   print(max)

# list_max = list_max(myList)

# 파이썬 3주차 과제 8차시 문제 2번
numbers = [42, 17, 23, 56, 9, 34]
def findMin(numList):
    min = numList[0]
    for num in numList:
        if num < min:
            min = num
    return min

result = findMin(numbers)
print(f"리스트의 최솟값은 {result}입니다.")
```
